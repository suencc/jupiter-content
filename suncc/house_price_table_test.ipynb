#### 房价统计

```python
import pymongo
def get_mongo_db():
    conn = pymongo.MongoClient('192.168.1.6',27017).house_info
    conn.authenticate('beta','beta')
    return conn
    
def count_house_old(db):
    hids = db.rsf_houses.distinct("hid")
    print(len(hids))
    for each in hids:
        xes = db.rsf_houses.find({"hid": each})
        total = 0;
        count = 0;
        for x in xes:
            total += x["house_price"]
            count += 1
        avg = int(total/count)
        if count > 1:
            print("multi:")
            print(each)
            print(avg)
        
        
```

#####  统计多房价格

```python
db = get_mongo_db()
count_house_old(db) 
```
