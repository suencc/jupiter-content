```python
# -*- coding:utf-8 -*-
import pymongo
from math import *

def get_monggo_db():
    conn = pymongo.MongoClient('127.0.0.1',27017).locachoice
    conn.authenticate('locachoice','locachoice')
    return conn

def get_result_grids_coll():
    conn = get_monggo_db()
    return conn['result_grids']

def clean_result_grids_coll():
    db = get_monggo_db()
    re = db.drop_collection('result_grids')
    print(re) 

# input lat_A 纬度A
# input lng_A 经度A
# input lat_B 纬度B
# input lng_B 经度B
# output distance 距离(km)
def calcDistance(lat_A, lng_A, lat_B, lng_B):
    try:
        if distances(lat_A, lat_B) and distances(lng_A, lng_B):
            return 0.001
        ra = 6378.140  # 赤道半径 (km)
        rb = 6356.755  # 极半径 (km)
        flatten = (ra - rb) / ra  # 地球扁率
        rad_lat_A = radians(lat_A)
        rad_lng_A = radians(lng_A)
        rad_lat_B = radians(lat_B)
        rad_lng_B = radians(lng_B)
        pA = atan(rb / ra * tan(rad_lat_A))
        pB = atan(rb / ra * tan(rad_lat_B))
        xx = acos(sin(pA) * sin(pB) + cos(pA) * cos(pB) * cos(rad_lng_A - rad_lng_B))
        c1 = (sin(xx) - xx) * (sin(pA) + sin(pB)) ** 2 / cos(xx / 2) ** 2
        c2 = (sin(xx) + xx) * (sin(pA) - sin(pB)) ** 2 / sin(xx / 2) ** 2
        dr = flatten / 8 * (c1 - c2)
        distance = ra * (xx + dr)
        return distance

    except Exception as e:
        print(e)
        return 0.001
    
def distances(a, b):
    a = a*1000000
    b = b*1000000
    if a>b:
        return (a-b)<=2
    else:
        return (b-a)<=2


START_LNG  = 113.297942
END_LNG = 113.466578
START_LAT = 23.093353
END_LAT = 23.263226
LNG_GRIDS = 20
LAT_GRIDS = 18
LNG_DELTA = round((END_LNG - START_LNG) / LNG_GRIDS, 7)
LAT_DELTA = round((END_LAT - START_LAT) / LAT_GRIDS, 7)

def create_result_grids():
    top_lat = END_LAT
    bottom_lat = top_lat
    coll = get_result_grids_coll()
    for row in range(1, LAT_GRIDS+1):
        bottom_lat = round(top_lat - LAT_DELTA, 7)
        left_lng = START_LNG
        right_lng = left_lng        
        for colum in range(1, LNG_GRIDS+1):
            right_lng = round(left_lng + LNG_DELTA, 7)
            doc ={
                'gridid': 'R%dC%d' % (row, colum),
                'left_lng': left_lng,
                'right_lng': right_lng,
                'top_lat': top_lat,
                'bottom_lat': bottom_lat,
            }            
            coll.insert_one(doc)
            left_lng = right_lng        
        top_lat = bottom_lat       

def append_cbd_distance2grids():
    cbd_lng = 113.32769
    cbd_lat = 23.125212
    coll = get_result_grids_coll()
    cursor = coll.find()
    for cur in cursor:
        lng = (cur.get('left_lng') + cur.get('right_lng')) / 2
        lat = (cur.get('top_lat') + cur.get('bottom_lat')) / 2
        cbd_distance = calcDistance(lat, lng, cbd_lat, cbd_lng)
        cbd_distance = round(cbd_distance, 2)
        print(cur.get('gridid'), cbd_distance)
        coll.update_one({'gridid': cur.get('gridid')}, {'$set': {'cbd_distance': cbd_distance}})
        
        
def test4grids():
    coll = get_result_grids_coll()    
    print('lat test print ......')
    for row in range(1, LAT_GRIDS+1):
        for colum in range(1, LNG_GRIDS+1):
            gridid = 'R%dC%d' % (row, colum)
            doc = coll.find_one({'gridid': gridid})
            rsltstr = '%3.8f %3.8f' % (doc.get('top_lat'), doc.get('bottom_lat'))
            print(rsltstr, end='|')
        print('')
    print('lng test print ......')
    for row in range(1, LAT_GRIDS+1):
        for colum in range(1, LNG_GRIDS+1):
            gridid = 'R%dC%d' % (row, colum)
            doc = coll.find_one({'gridid': gridid})
            rsltstr = '%3.8f %3.8f' % (doc.get('left_lng'), doc.get('right_lng'))
            print(rsltstr, end='|')
        print('')    
        
if __name__ == '__main__':
#     create_result_grids()
#     test4grids()
    append_cbd_distance2grids()
```
