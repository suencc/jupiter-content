```python
# -*- coding:utf-8 -*-
import pymongo
import locachoice_utils

def get_monggo_db():
    conn = pymongo.MongoClient('127.0.0.1',27017).locachoice
    conn.authenticate('locachoice','locachoice')
    return conn

def get_crawl_coll():
    conn = get_monggo_db()
    return conn['crawl_navinfo_road']

def get_etl_coll():
    conn = get_monggo_db()
    return conn['etl_navinfo_road']

def clean_etl_coll():
    db = get_monggo_db()
    re = db.drop_collection("etl_navinfo_road")
    print(re)

def get_tmp_coll():
    conn = get_monggo_db()
    return conn['tmp_navinfo_road']

def clean_tmp_coll():
    db = get_monggo_db()
    re = db.drop_collection("tmp_navinfo_road")
    print(re)

def get_result_coll():
    conn = get_monggo_db()
    return conn['result_navinfo_road']

def clean_result_coll():
    db = get_monggo_db()
    re = db.drop_collection("result_navinfo_road")
    print(re)

def get_result_grids():
    conn = get_monggo_db()
    return conn['result_grids']   

def get_id_from_desc(desc):
    index = desc.find('-')
    ret = desc
    if index == -1:
        pass 
    else:
        ret = desc[:index]
    return ret
#     
def export_crawl2tmp():
    src_coll = get_crawl_coll()
    clean_tmp_coll()
    dst_coll = get_tmp_coll()
    cursor = src_coll.find()
    for cur in cursor:
        doc = {
            'id': get_id_from_desc(cur.get('id')),
            'name' : cur.get('name'),
            'direction':cur.get('id')[6],
            'startName' : cur.get('startName'),
            'endName' : cur.get('endName'),  
            'avgspeed' : float(cur.get('avgspeed')),
            'cIndex' : float(cur.get('cIndex')),
            'sIndex' : float(cur.get('sIndex')),
            'bIndex' : float(cur.get('bIndex')),
            'time' : cur.get('Nowdate')[11:],           
        }
        dst_coll.insert_one(doc)
    
def export_tmp2etl():
    src_coll = get_tmp_coll()    
    clean_etl_coll()
    dst_coll = get_etl_coll()
    
    times = src_coll.distinct('time')
    ids = src_coll.distinct('id')
    for time in times:
        print('one time is procing')
        for id in ids:            
            total_avgspeed = 0
            total_cindex = 0
            total_sindex = 0
            total_bindex = 0
            count = 0
            cursor = src_coll.find({'id': id, 'time': time})
            for cur in cursor:
                count += 1
                total_avgspeed += float(cur.get('avgspeed'))
                total_cindex += float(cur.get('cIndex'))
                total_sindex += float(cur.get('sIndex'))
                total_bindex += float(cur.get('bIndex'))            
            print('count: ', count)
            if count == 0:
                continue
            avg_avgspeed = round(total_avgspeed / count, 2)
            avg_cindex = round(total_cindex / count, 2)
            avg_sindex = round(total_sindex / count, 2)
            avg_bindex = round(total_bindex / count, 2)
            
            src_doc = src_coll.find_one({'id': id, 'time': time})
            doc = {
                'id': id,
                'name': src_doc.get('name'),
                'direction':  src_doc.get('direction'),
                'startName' : src_doc.get('startName'),
                'endName' : src_doc.get('endName'), 
                'avg_avgspeed': avg_avgspeed,
                'avg_cindex': avg_cindex,
                'avg_sindex': avg_sindex,
                'avg_bindex': avg_bindex,
                'time':time,
            }
            dst_coll.insert_one(doc)
            
# 时段划分，全按闭区间处理
# pre_work 06:00~08:00
# go_work 08:00~10:00
# go_diner 17:30~19:30
# go_home 19:30~22:30
def export_etl2result():
    src_coll = get_etl_coll()    
    clean_result_coll()
    dst_coll = get_result_coll()
    ids = src_coll.distinct('id')
    for id in ids:
        print('is statits one id...')
        # pre_work        
        total_avgspeed = 0
        total_cindex = 0
        total_sindex = 0
        total_bindex = 0
        count = 0        
        cursor = src_coll.find({'id': id, 'time': {'$gte': '06:00', '$lte': '08:00'}})
        for cur in cursor:
            count += 1
            total_avgspeed += float(cur.get('avg_avgspeed'))
            total_cindex += float(cur.get('avg_cindex'))
            total_sindex += float(cur.get('avg_sindex'))
            total_bindex += float(cur.get('avg_bindex'))        
        
        try:
            avg_avgspeed = round(total_avgspeed / count, 2)
            avg_cindex = round(total_cindex / count, 2)
            avg_sindex = round(total_sindex / count, 2)
            avg_bindex = round(total_bindex / count, 2) 
        except Exception as e:
            avg_avgspeed = 0
            avg_cindex = 0
            avg_sindex = 0
            avg_bindex = 0             
        
        src_doc = src_coll.find_one({'id': id, 'time': {'$gte': '06:00', '$lte': '08:00'}})
        doc = {
            'id': id,
            'name': src_doc.get('name'),
            'direction':  src_doc.get('direction'),
            'startName' : src_doc.get('startName'),
            'endName' : src_doc.get('endName'), 
            'avg_avgspeed': avg_avgspeed,
            'avg_cindex': avg_cindex,
            'avg_sindex': avg_sindex,
            'avg_bindex': avg_bindex,
            'timeslot':'pre_work',
        }
        dst_coll.insert_one(doc)         
            
        # go_work        
        total_avgspeed = 0
        total_cindex = 0
        total_sindex = 0
        total_bindex = 0
        count = 0        
        cursor = src_coll.find({'id': id, 'time': {'$gte': '08:00', '$lte': '10:00'}})
        for cur in cursor:
            count += 1
            total_avgspeed += float(cur.get('avg_avgspeed'))
            total_cindex += float(cur.get('avg_cindex'))
            total_sindex += float(cur.get('avg_sindex'))
            total_bindex += float(cur.get('avg_bindex'))        
        
        try:
            avg_avgspeed = round(total_avgspeed / count, 2)
            avg_cindex = round(total_cindex / count, 2)
            avg_sindex = round(total_sindex / count, 2)
            avg_bindex = round(total_bindex / count, 2) 
        except Exception as e:
            avg_avgspeed = 0
            avg_cindex = 0
            avg_sindex = 0
            avg_bindex = 0             
        
        src_doc = src_coll.find_one({'id': id, 'time': {'$gte': '08:00', '$lte': '10:00'}})
        doc = {
            'id': id,
            'name': src_doc.get('name'),
            'direction':  src_doc.get('direction'),
            'startName' : src_doc.get('startName'),
            'endName' : src_doc.get('endName'), 
            'avg_avgspeed': avg_avgspeed,
            'avg_cindex': avg_cindex,
            'avg_sindex': avg_sindex,
            'avg_bindex': avg_bindex,
            'timeslot':'go_work',
        }
        dst_coll.insert_one(doc)
        
        # go_diner        
        total_avgspeed = 0
        total_cindex = 0
        total_sindex = 0
        total_bindex = 0
        count = 0        
        cursor = src_coll.find({'id': id, 'time': {'$gte': '17:30', '$lte': '19:30'}})
        for cur in cursor:
            count += 1
            total_avgspeed += float(cur.get('avg_avgspeed'))
            total_cindex += float(cur.get('avg_cindex'))
            total_sindex += float(cur.get('avg_sindex'))
            total_bindex += float(cur.get('avg_bindex'))        
        
        try:
            avg_avgspeed = round(total_avgspeed / count, 2)
            avg_cindex = round(total_cindex / count, 2)
            avg_sindex = round(total_sindex / count, 2)
            avg_bindex = round(total_bindex / count, 2) 
        except Exception as e:
            avg_avgspeed = 0
            avg_cindex = 0
            avg_sindex = 0
            avg_bindex = 0             
        
        src_doc = src_coll.find_one({'id': id, 'time': {'$gte': '17:30', '$lte': '19:30'}})
        doc = {
            'id': id,
            'name': src_doc.get('name'),
            'direction':  src_doc.get('direction'),
            'startName' : src_doc.get('startName'),
            'endName' : src_doc.get('endName'), 
            'avg_avgspeed': avg_avgspeed,
            'avg_cindex': avg_cindex,
            'avg_sindex': avg_sindex,
            'avg_bindex': avg_bindex,
            'timeslot':'go_diner',
        }
        dst_coll.insert_one(doc)   
        
        # go_home        
        total_avgspeed = 0
        total_cindex = 0
        total_sindex = 0
        total_bindex = 0
        count = 0        
        cursor = src_coll.find({'id': id, 'time': {'$gte': '19:30', '$lte': '22:30'}})
        for cur in cursor:
            count += 1
            total_avgspeed += float(cur.get('avg_avgspeed'))
            total_cindex += float(cur.get('avg_cindex'))
            total_sindex += float(cur.get('avg_sindex'))
            total_bindex += float(cur.get('avg_bindex'))        
        
        try:
            avg_avgspeed = round(total_avgspeed / count, 2)
            avg_cindex = round(total_cindex / count, 2)
            avg_sindex = round(total_sindex / count, 2)
            avg_bindex = round(total_bindex / count, 2) 
        except Exception as e:
            avg_avgspeed = 0
            avg_cindex = 0
            avg_sindex = 0
            avg_bindex = 0             
        
        src_doc = src_coll.find_one({'id': id, 'time': {'$gte': '19:30', '$lte': '22:30'}})
        doc = {
            'id': id,
            'name': src_doc.get('name'),
            'direction':  src_doc.get('direction'),
            'startName' : src_doc.get('startName'),
            'endName' : src_doc.get('endName'), 
            'avg_avgspeed': avg_avgspeed,
            'avg_cindex': avg_cindex,
            'avg_sindex': avg_sindex,
            'avg_bindex': avg_bindex,
            'timeslot':'go_home',
        }
        dst_coll.insert_one(doc)
        
def append_location2result():
    coll = get_result_coll()
    ids = coll.distinct('id')
    for id in ids:    
        cur = coll.find_one({'id': id})
        toset_doc ={}
        startaddr = '%s和%s路口' % (cur.get('name'), cur.get('startName'))
        endaddr = '%s和%s路口' % (cur.get('name'), cur.get('endName'))
        start_location = locachoice_utils.getlocation(startaddr)
        if isinstance(start_location, int):
            print('can not get location:', startaddr)
            toset_doc['start_lng'] = 0
            toset_doc['start_lat'] = 0
        else:
            toset_doc['start_lng'] = round(start_location.get('lng'), 7)
            toset_doc['start_lat'] = round(start_location.get('lat'), 7)

        end_location = locachoice_utils.getlocation(endaddr)
        if isinstance(end_location, int):
            print('can not get location:', endaddr)
            toset_doc['end_lng'] = 0
            toset_doc['end_lat'] = 0
        else:
            toset_doc['end_lng'] = round(end_location.get('lng'), 7)
            toset_doc['end_lat'] = round(end_location.get('lat'), 7)

        toset_doc['mid_lng'] = round((toset_doc.get('start_lng')+toset_doc.get('end_lng'))/2, 7)
        toset_doc['mid_lat'] = round((toset_doc.get('start_lat')+toset_doc.get('end_lat'))/2, 7)
        toset_doc['gridid'] = locachoice_utils.get_grid_id(toset_doc.get('mid_lng'), toset_doc.get('mid_lat'))
        print(toset_doc)
        coll.update_many({'id': id}, {'$set': toset_doc}) 
# 一般不需要调用，纠正 append_location2result 中的bug       
def update_gridid_in_result():
    coll = get_result_coll()
    cursor = coll.find()
    for cur in cursor:
        gridid = locachoice_utils.get_grid_id(cur.get('mid_lng'), cur.get('mid_lat'))
        print(gridid)
        coll.update_one({'_id': cur.get('_id')}, {'$set': {'gridid': gridid}})
        
def fill_navinfo_fields_in_grids():
    grids_coll = get_result_grids()
    cursor = grids_coll.find({'speed_pre_work': {'$exists': False}})
    for cur in cursor:
        toset_doc = {
            'speed_pre_work': None,
            'cindex_pre_work': None,
            'speed_go_work': None,
            'cindex_go_work': None,
            'speed_go_diner': None,
            'cindex_go_diner': None,
            'speed_go_home': None,
            'cindex_go_home': None,
        }
        print(toset_doc)
        grids_coll.update_one({'gridid': cur.get('gridid')}, {'$set': toset_doc})
                              

def export_navinfo2grids():
    coll = get_result_coll()
    grids_coll = get_result_grids()
    gridids = coll.distinct('gridid')
    for gridid in gridids:    
        if gridid == -1:
            continue
        #pre_work
        t_speed = 0 
        t_cindex = 0     
        count = 0
        cursor = coll.find({'gridid': gridid, 'timeslot': 'pre_work'})
        for cur in cursor:
            count += 1
            t_speed += cur.get('avg_avgspeed')
            t_cindex += cur.get('avg_cindex')
        speed_pre_work = round(t_speed/count, 2)
        cindex_pre_work = round(t_cindex/count, 2)
        #go_work
        t_speed = 0 
        t_cindex = 0     
        count = 0
        cursor = coll.find({'gridid': gridid, 'timeslot': 'go_work'})
        for cur in cursor:
            count += 1
            t_speed += cur.get('avg_avgspeed')
            t_cindex += cur.get('avg_cindex')
        speed_go_work = round(t_speed/count, 2)
        cindex_go_work = round(t_cindex/count, 2)
        #go_diner
        t_speed = 0 
        t_cindex = 0     
        count = 0
        cursor = coll.find({'gridid': gridid, 'timeslot': 'go_diner'})
        for cur in cursor:
            count += 1
            t_speed += cur.get('avg_avgspeed')
            t_cindex += cur.get('avg_cindex')
        speed_go_diner = round(t_speed/count, 2)
        cindex_go_diner = round(t_cindex/count, 2)
        #go_home
        t_speed = 0 
        t_cindex = 0     
        count = 0
        cursor = coll.find({'gridid': gridid, 'timeslot': 'go_home'})
        for cur in cursor:
            count += 1
            t_speed += cur.get('avg_avgspeed')
            t_cindex += cur.get('avg_cindex')
        speed_go_home = round(t_speed/count, 2)
        cindex_go_home = round(t_cindex/count, 2)
        toset_doc = {
            'speed_pre_work': speed_pre_work,
            'cindex_pre_work': cindex_pre_work,
            'speed_go_work': speed_go_work,
            'cindex_go_work': cindex_go_work,
            'speed_go_diner': speed_go_diner,
            'cindex_go_diner': cindex_go_diner,
            'speed_go_home': speed_go_home,
            'cindex_go_home': cindex_go_home,
        }
        print(toset_doc)
        grids_coll.update_one({'gridid': gridid}, {'$set': toset_doc})
    fill_navinfo_fields_in_grids()
    
    
if __name__ == '__main__':
    print('starting...')
#     export_navinfo2grids()
#     fill_navinfo_fields_in_grids()
    print('ended ended')
```
